```
print("Su pedido no puede pasar los 10,000gr entre ambos productos")
print("Peso del payaso: 150gr")
print("Peso del juego de mesa: 170gr")
def pedido(payasos, pesopayaso, juegos, pesojuego):
  pedido_payaso = payasos * pesopayaso
  Pedido_juego = juegos * pesojuego
  total = pedido_payaso + Pedido_juego
  if total < 100000:
    print("Operacion exitosa")
    print("Su peido en gr es: ",total)
  else:
    print("Error")
    sobre_peso = total - 10000
    print("Su pedido se excedio por: ",sobre_peso,"gr")
p = int(input("Numero de payasos: "))
pesopayaso = 150
j = int(input("Numero de juegos: "))
pesojuego = 170
pedido(p, pesopayaso, j, pesojuego)
```
